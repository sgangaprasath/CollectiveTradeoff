```python
class params():
    def __init__(self) -> None:
        super().__init__()
        self.nMax = 200
        self.dr = 0.5
        self.rxInit = 0.
        self.ryInit = 0.
        self.thetaInit = 0.
        self.rSt = 6.0
        self.eps = 0.25
        self.actMax = 2*np.pi
        self.rng = random.PRNGKey(0)
```

```python
class antMotor():
    def __init__(self, r, thet, rng):
        super().__init__()
        self.r = r
        self.thet = thet
        self.rng = rng
    
    @jit
    def stepFwd(self, dr):
        drng = random.split(self.rng)[0]
        new_thet = self.thet + random.uniform(drng, minval=-jnp.pi/4, maxval=jnp.pi/4)
        new_r = self.r + dr*jnp.array([jnp.cos(self.thet), jnp.sin(self.thet)])
        return antMotor(new_r, new_thet, drng)
    
    @jit
    def homing(self, dr, new_thet):
        drng = random.split(self.rng)[0]
        new_r = self.r + dr*jnp.array([jnp.cos(new_thet), jnp.sin(new_thet)])  
        return antMotor(new_r, new_thet, drng)
    
register_pytree_node(antMotor,
                     lambda x: ((x.r[0], x.r[1], x.thet, x.rng), None),
                     lambda _, tup: antMotor(jnp.array([tup[0], tup[1]]), tup[2], tup[3]))
```

```python
class antDyn(gym.Env):
    def __init__(self):
        super(antDyn, self).__init__()
        # Parameters
        self.nSteps = 0
        self.dr = p.dr
        self.rxInit = p.rxInit
        self.ryInit = p.ryInit
        self.thetaInit = p.thetaInit
        self.rSt = p.rSt
        self.actMax = p.actMax
        self.nMax = p.nMax
        self.theta0 = p.thetaInit
        self.rx0 = p.rxInit
        self.ry0 = p.ryInit
        self.rng = p.rng
        self.dr = p.dr
        # Variables in the model
        self.rx = [self.rx0]
        self.ry = [self.ry0]
        self.theta = [self.theta0]

        self.observation_space = spaces.Box(low=np.array([-10., -10., 0]),
                                            high=np.array([10., 10., 2*np.pi]),
                                            dtype=np.float32)
        self.action_space = spaces.Box(low=np.array([0.]), high=np.array([self.actMax]))
        
        self.a_history = []
    
    def propgtr(self, act):
        self.rng = random.split(self.rng)[0]
        new_theta = self.theta[-1] + random.uniform(self.rng, minval=-jnp.pi/4, maxval=jnp.pi/4)
        new_rx = self.rx[-1] + self.dr*(jnp.cos(self.theta[-1]) + jnp.cos(act))
        new_ry = self.ry[-1] + self.dr*jnp.sin(self.theta[-1] + jnp.sin(act))
        return jnp.array([new_rx, new_ry, new_theta])
        # theta, thetaDt, phi, phiDt, xc = y
        # return np.array([thetaDt, u1*np.sin(theta) - u2*(theta+phi) - u3*thetaDt + act, phiDt, - w1*(theta+phi)- w2*phiDt + act, rw*phiDt]).astype(np.float32)
    
    def step(self, action):
        a = action[0]
        self.a_history.append(a)
        sol = self.propgtr(a)
        rx = sol[0]
        self.rx.append(rx)
        ry = sol[1]
        self.ry.append(ry)
        theta = np.mod(sol[2], 2*np.pi)
        self.theta.append(theta)
        
        state = np.array([self.rx[-1], self.ry[-1], self.theta[-1]]).astype(np.float32)
        if self.nSteps < (self.nMax):
            reward = - np.sqrt((rx-self.rSt)**2 + ry**2)
            # (self.lu*np.cos(theta[0])-self.lu*np.cos(self.thetaSt))**2 - (self.lu*np.cos(phi[0])-self.lu*np.cos(self.phiSt))**2
            done = False
        else:
            reward = -1.0
            done = True
        info = {}
        self.nSteps += 1
        return state, reward, done, info
    
    def reset(self):
        self.nSteps = 0
        
        self.rx = [self.rx0]
        self.ry = [self.ry0]
        self.theta = [self.theta0]
        
        state = np.array([self.rx0, self.ry0, self.theta0]).astype(np.float32)
        return state
    
    def render(self):
        pass
    
    def close(self):
        pass
```

```python
p = params()
env = antDyn()
model = PPO("MlpPolicy", env, verbose=1)
model.learn(total_timesteps=5e5)
```

```python
obs = env.reset()
done = False
while not done:
    action, _states = model.predict(obs, deterministic=True)
    obs, reward, done, info = env.step(action)
    env.render()

a_actions = copy.deepcopy(env.a_history)
a_actions.insert(0, 0)
```
